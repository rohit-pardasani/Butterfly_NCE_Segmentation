```python
import cv2
import json
import numpy as np
import os
countour_df = pd.read_csv('./SegmentedDataset/mask_contours_train_val.csv')

# Create output directory for masks
output_dir = './binary_masks_train_val/'
os.makedirs(output_dir, exist_ok=True)

# Iterate through the dataframe rows
for index, row in countour_df.iterrows():
    # Parse the polygon coordinates
    shape_attributes = json.loads(row['region_shape_attributes'])
    all_points_x = shape_attributes['all_points_x']
    all_points_y = shape_attributes['all_points_y']
    
    # Create a blank mask
    mask = np.zeros((224, 224), dtype=np.uint8)  # Assuming 256x256 image size, adjust as needed
    
    # Create a polygon from the points
    points = np.array(list(zip(all_points_x, all_points_y)), dtype=np.int32)
    cv2.fillPoly(mask, [points], color=255)
    
    # Save the mask as a .png file
    mask_filename = os.path.join(output_dir, f"{row['filename'].split('.')[0]}_mask.png")
    cv2.imwrite(mask_filename, mask)

print(f"Binary masks saved in {output_dir}")
```
